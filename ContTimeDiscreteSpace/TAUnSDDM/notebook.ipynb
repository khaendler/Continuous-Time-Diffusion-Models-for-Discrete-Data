```python
import jax.numpy as jnp
import torch
import jax.random as jrandom
```

```python
key = jrandom.PRNGKey(0)
xt = jrandom.uniform(key, (4, 3, 6))

print()
qt = jnp.expand_dims(jnp.arange(xt.shape[0]), tuple(range(1, xt.ndim)))
print(qt, qt.shape)

print(jnp.expand_dims(xt, axis=list(range(1, 4))).shape)
print(xt,jnp.expand_dims(xt, axis=list(range(1, 4))))
```

```python
def expand_dims(x, axis):
    for i in axis:
        x = x.unsqueeze(i)
    return x
```

```python


# Schritt 2: Erzeugen Sie qt
xt = torch.rand(4, 3, 6)
print(expand_dims(torch.arange(xt.shape[0]), tuple(range(1, xt.ndim))).shape)
print(expand_dims(xt, list(range(1, 4))).shape)
qt_expanded = torch.arange(xt.shape[0]).reshape(tuple([-1] + [1]*(xt.ndim-1)))
#print(torch.arange(xt.shape[0]).expand(*(range(1, xt.ndim))))
print(qt_expanded)
print(qt_expanded.shape)

# Verwenden Sie die `reshape` Methode mit den vorher definierten Größen
result = xt.reshape(*[xt.shape[0]] + [1 for _ in range(1, xt.ndim)] + list(xt.shape[1:]))
print(result.shape)
print(xt, result)
```
