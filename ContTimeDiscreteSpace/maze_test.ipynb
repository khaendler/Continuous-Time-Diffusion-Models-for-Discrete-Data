```python
from datasets.datasets import get_maze_data
from tauLDR.config.config_train_sample import get_config 
import torch
```

```python
config = get_config()
maze_data = get_maze_data(config)

print(torch.unique(next(iter(maze_data))))
print(next(iter(maze_data)).shape)
```

```python
import matplotlib.pyplot as plt

# Eine Funktion, um einige Bilder zu plotten
def show_images(images, n=8):
    plt.figure(figsize=(15, 15))
    for i in range(n):
        ax = plt.subplot(1, n, i + 1)
        plt.imshow(images[i].permute(1, 2, 0).numpy().astype("uint8"), cmap='Greys')
        plt.axis("off")
    plt.show()

# Nehmen Sie einen Batch von Bildern
images = next(iter(maze_data))
show_images(images)
```
