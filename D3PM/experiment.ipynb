```python
import numpy as np
import torch
from networks import UNet
```

```python

unet = UNet(in_channel=1,out_channel=1,
            channel=32,
            channel_multiplier=[1, 2, 2, 2],
            n_res_blocks=1,
            attn_resolutions=[16],
            num_heads=1,
            dropout=0,
            model_output='logits',  # 'logits' or 'logistic_pars'
            num_pixel_vals=256,
            img_size=32)
```

```python
x = torch.randn(16, 1, 32, 32)
batch_size = 16
channels = 1
height = 32
width = 32
x = torch.randint(0, 256, (batch_size, channels, height, width), dtype=torch.uint8)

t = (torch.randint(low=0, high=(10), size=(16,)))
output_tensor = unet(x, t)
print("Output shape:", output_tensor.shape)
total_params = sum(p.numel() for p in unet.parameters())

print(total_params)
```
